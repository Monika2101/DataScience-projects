```python
data_1 = pd.get_dummies(data[['CRIME']])
data_1.head()
```

```python
data.drop(['CRIME'],axis=1, inplace=True)
```

```python
data_f =pd.concat([data, data_1], axis=1)
```

```python
data_f.head()
```

```python
data_f.set_index('STATE')['CRIME_Rape'].plot(kind = 'line', figsize = (12,8))
plt.xlabel('STATE')
plt.ylabel('No. of Cases')
plt.title('CRIME_Rape Cases')
plt.show()
```
