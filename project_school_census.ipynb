```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

#finding null value

```python
data.isnull().sum()
```

```python
data.drop(data.iloc[:, 64:78], axis=1 , inplace=True)
```

#Removing columns with null values

```python
data.drop(["Other ethnicity","Wrist circumference","Bag weight","Screen time opinion - On your phone","Screen time opinion - On social media"], axis=1 , inplace=True)
```

```python
data.head()
```

```python
data.isnull().sum()
```

```python
data.drop(["Check messages as soon as you wake up", "Respond to messages immediately", "Take phone to school",
                              "Lose focus as school due to phone", "Bed time", "Wake time","Time you get home from school","Time you ate dinner"],axis=1, inplace=True)
```

```python
data.isnull().sum()
```

```python
data.head()
```

```python
data["Physical activity - before school"] = data["Physical activity - before school"].str.split('+').str[0]
```

```python
data["Physical activity - at school"] = data["Physical activity - at school"].str.split('+').str[0]
```

```python
data["Physical activity - after school"] = data["Physical activity - after school"].str.split('+').str[0]
```

```python
data.iloc[:,26:28] = data.iloc[:,26:28].astype(float)
```

```python
data_object = data.select_dtypes(include=['object'])
```

```python
data_object.head()
```

```python
data_object.isnull().sum()
```

```python
data_object = data_object.drop(['Favourite video game'], axis=1)
```

```python
data_nu= data._get_numeric_data()
```

```python
data_nu.fillna(data_nu.mean(), inplace = True)
```

```python
data_nu.head()
```

```python
data_nu.isnull().sum()
```

```python
census_final_data = pd.concat([data_nu,data_object],axis=1)
```

```python
census_final_data.head()
```

```python
census_final_data=census_final_data.dropna()
```

```python
census_final_data.head()
```

```python
census_final_data = census_final_data.replace('n/a', np.nan)
```
